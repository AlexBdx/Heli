```python
import numpy as np
import matplotlib.pyplot as plt
speed_light = 300000000
```

```python
f = 94*10**9
Tc = 100*10**-6
l = speed_light/f
v_max = l/(4*Tc)
print("Elva-1: {:.2f} m/s".format(v_max))
```

```python
f = 77*10**9
Tc = 40*10**-6
l = speed_light/f
v_max = l/(4*Tc)
print("TI : {:.2f} m/s".format(v_max))
```

```python
center_freq = np.linspace(1, 25, 100)*10**9
plt.figure(figsize=(10, 5))
list_v_max = [10, 20, 50, 100, 200, 500]
for v_max in list_v_max:
    chirp_freq = (4/speed_light)*center_freq*v_max
    plt.scatter(center_freq/10**9, chirp_freq/1000, s=2)
#plt.axis([10**9, 10**11, 0, 0.01])
plt.legend([str(v) for v in list_v_max])
plt.plot([5.3, 5.3], [0, 50], '--')
plt.xlabel("Center frequency [GHz]")
plt.ylabel("Required Tc [kHz]")
plt.grid(True)
plt.title("Required Tc to meet the target vmax")
```
